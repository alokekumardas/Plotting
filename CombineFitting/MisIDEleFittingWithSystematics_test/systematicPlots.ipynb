```python
# sampleList = ['TTGamma', 'TTbar', 'TGJets','SingleTop', 'WJets', 'ZJets', 'WGamma','ZGamma','Diboson','TTV','GJets',"QCD"]
sampleList =["ZJets"]
path = " /uscms_data/d3/npoudyal/TTGammaSemiLeptonic13TeV/CMSSW_10_2_14/src/TTGammaSemiLep_13TeV/Plotting/"
```

```python
%jsroot
canvas = ROOT.TCanvas()
for ifile in sampleList:
    myfile   = ROOT.TFile(path+"histograms_2016/ele/hists_looseCRge2e0/"+ifile+".root","read")
    myfileUp = ROOT.TFile(path+"histograms_2016/ele/hists_BTagSF_b_up_looseCRge2e0/"+ifile+".root","read")
    myfileDo = ROOT.TFile(path+"histograms_2016/ele/hists_BTagSF_b_down_looseCRge2e0/"+ifile+".root","read")
    hist   = myfile.  Get("phosel_MassEGamma_%s"%ifile); hist  .SetLineColor(ROOT.kRed)
    histUp = myfileUp.Get("phosel_MassEGamma_%s"%ifile); histUp.SetLineColor(ROOT.kBlue)
    histDo = myfileDo.Get("phosel_MassEGamma_%s"%ifile); histDo.SetLineColor(ROOT.kGreen)
    hist.  Draw("hist")
    histUp.Draw("hist same")
    histDo.Draw("hist same")
    canvas.Draw()
```
