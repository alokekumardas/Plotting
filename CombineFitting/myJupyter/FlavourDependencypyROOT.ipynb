```python
import numpy, sys
import ROOT
import matplotlib.pyplot as plt
```

```python
#fileName = '/uscms_data/d3/npoudyal/TTGammaSemiLeptonic13TeV/CMSSW_10_2_14/src/Danny_TTGammaSemiLep_13TeV/TTGammaSemiLep_13TeV/WGamma_01J_5f_2016_AnalysisNtuple.root'
fileName = '/uscms_data/d3/npoudyal/TTGammaSemiLeptonic13TeV/CMSSW_10_2_14/src/Danny_TTGammaSemiLep_13TeV/TTGammaSemiLep_13TeV/ZGamma_01J_5f_lowMass_2016_AnalysisNtuple.root'

myfile = ROOT.TFile(fileName,'read')
if myfile.IsZombie(): 
    print ("corrupted file",myfile)
    sys.exit()
mytree = myfile.AnalysisTree
```

```python
lightjet = 0
bjet     = 0
cjet     = 0

line=''
for iev in mytree:
    lepCut = iev.passPresel_Ele
    phoCut = iev.nPho
    jetCut = iev.nJet
    if lepCut and phoCut==1 and jetCut >= 2: 
        for ijet in xrange(iev.nJet):
            if iev.jetHadronFlavour[ijet]==0:
                lightjet+=1
            if iev.jetHadronFlavour[ijet]==4:
                cjet+=1
            if iev.jetHadronFlavour[ijet]==5:
                bjet+=1

line += "\\newcommand{\\%s} {\ensuremath{%i }}\n"%("crAlllightjetEleZG",lightjet)
line += "\\newcommand{\\%s} {\ensuremath{%i }}\n"%("crAllcjetEleZG",cjet)
line += "\\newcommand{\\%s} {\ensuremath{%i }}\n"%("crAllbjetEleZG",bjet)

lightjet = 0
bjet     = 0
cjet     = 0
for iev in mytree:
    lepCut = iev.passPresel_Ele
    phoCut = iev.nPho
    jetCut = iev.nJet
    if lepCut and phoCut==1 and jetCut == 2: 
        for ijet in xrange(iev.nJet):
            if iev.jetHadronFlavour[ijet]==0:
                lightjet+=1
            if iev.jetHadronFlavour[ijet]==4:
                cjet+=1
            if iev.jetHadronFlavour[ijet]==5:
                bjet+=1

line += "\\newcommand{\\%s} {\ensuremath{%i }}\n"%("crOnelightjetEleZG",lightjet)
line += "\\newcommand{\\%s} {\ensuremath{%i }}\n"%("crOnecjetEleZG",cjet)
line += "\\newcommand{\\%s} {\ensuremath{%i }}\n"%("crOnebjetEleZG",bjet)

lightjet = 0
bjet     = 0
cjet     = 0

for iev in mytree:
    lepCut = iev.passPresel_Ele
    phoCut = iev.nPho
    jetCut = iev.nJet
    if lepCut and phoCut==1 and jetCut == 3: 
        for ijet in xrange(iev.nJet):
            if iev.jetHadronFlavour[ijet]==0:
                lightjet+=1
            if iev.jetHadronFlavour[ijet]==4:
                cjet+=1
            if iev.jetHadronFlavour[ijet]==5:
                bjet+=1

line += "\\newcommand{\\%s} {\ensuremath{%i }}\n"%("crTwolightjetEleZG",lightjet)
line += "\\newcommand{\\%s} {\ensuremath{%i }}\n"%("crTwocjetEleZG",cjet)
line += "\\newcommand{\\%s} {\ensuremath{%i }}\n"%("crTwobjetEleZG",bjet)

lightjet = 0
bjet     = 0
cjet     = 0

for iev in mytree:
    lepCut = iev.passPresel_Ele
    phoCut = iev.nPho
    jetCut = iev.nJet
    if lepCut and phoCut==1 and jetCut >= 4: 
        for ijet in xrange(iev.nJet):
            if iev.jetHadronFlavour[ijet]==0:
                lightjet+=1
            if iev.jetHadronFlavour[ijet]==4:
                cjet+=1
            if iev.jetHadronFlavour[ijet]==5:
                bjet+=1

line += "\\newcommand{\\%s} {\ensuremath{%i }}\n"%("crThreelightjetEleZG",lightjet)
line += "\\newcommand{\\%s} {\ensuremath{%i }}\n"%("crThreecjetEleZG",cjet)
line += "\\newcommand{\\%s} {\ensuremath{%i }}\n"%("crThreebjetEleZG",bjet)

lightjet = 0
bjet     = 0
cjet     = 0
for iev in mytree:
    lepCut = iev.passPresel_Ele
    phoCut = iev.nPho
    jetCut = iev.nJet
    if lepCut and phoCut==1 and jetCut == 2 and iev.nBJet==1: 
        for ijet in xrange(iev.nJet):
            if iev.jetHadronFlavour[ijet]==0:
                lightjet+=1
            if iev.jetHadronFlavour[ijet]==4:
                cjet+=1
            if iev.jetHadronFlavour[ijet]==5:
                bjet+=1

line += "\\newcommand{\\%s} {\ensuremath{%i }}\n"%("crFourlightjetEleZG",lightjet)
line += "\\newcommand{\\%s} {\ensuremath{%i }}\n"%("crFourcjetEleZG",cjet)
line += "\\newcommand{\\%s} {\ensuremath{%i }}\n"%("crFourbjetEleZG",bjet)



with open("FlavourMultiplicity_ele_ZG.tex","w") as _file:
    _file.write(line)

    
```
